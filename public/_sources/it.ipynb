# Programmeren 101

De site kan ook gebruikt worden om leerlingen in te leren programmeren in uiterst vriendelijke programmeertaal **PYTHON**.
Hier onder een voorbeeld:

```python
# Importeer de pakketten die moeten worden gebruikt
import matplotlib.pyplot as plt
import numpy as np

# Definieer x
x_beginwaarde = -1
x_eindwaarde = 2
stapgrootte = 0.01
x = np.arange(x_beginwaarde, x_eindwaarde, stapgrootte)

# Defineer y
y = 1 + 3*x

# Maak een figuur aan
fig, ax = plt.subplots()

# Plot x, y
ax.plot(x, y)

# Defineer wat er op de x-as en y-as moet staan
ax.set(xlabel='tijd (s)', ylabel='voltage (mV)')

# Defineer de titel van de figuur
ax.set(title='Zo simpel kan het zijn!')

# Set het achtergrond raster aan
ax.grid()
```
